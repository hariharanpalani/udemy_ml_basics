```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def find_cof(x, y):
    n = np.size(x)
    
    #mean values of x and y
    mean_x, mean_y = np.mean(x), np.mean(y)
    
    #calculation cross deviations
    Sxy = np.sum(x * y) - n * mean_x * mean_y
    Sxx = np.sum(x * x) - n * mean_x * mean_x
    
    b1 = Sxy / Sxx
    
    b0 = mean_y - b1 * mean_x
    
    return (b0, b1)
        
```

```python
def plt_regression_line(x, y, b):
    plt.scatter(x, y, color="m", marker="o", s=30)
    
    y_pred = b[0]+b[1]*x
    
    plt.plot(x, y_pred, color="g")
    plt.xlabel("X")
    plt.ylabel("Y")
    
    plt.show()
```

```python
#main functions
def main():
    tip = sns.load_dataset('tips')
    df = pd.DataFrame(tip)
    print(df.head())
    
    #observations
    x = df['total_bill']
    y = df['tip']
    
    #call find coefficient function
    b = find_cof(x, y)
    plt_regression_line(x, y, b)
    
if __name__=="__main__":
    main()
```
