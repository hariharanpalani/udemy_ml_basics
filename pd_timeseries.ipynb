```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fd = pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv', index_col='Date', parse_dates=True)
fd.head()
```

```python
fd.plot()
plt.show()
```

```python
df_timeseries = fd.groupby(fd.index.time).mean()
# range of 6 used to mention days of week
hx_series = 4 * 60 * 60 * np.arange(6)
df_timeseries.plot(xticks=hx_series)
plt.show()
```

```python
df_dwseries = fd.groupby(fd.index.dayofweek).mean()
df_dwseries.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df_dwseries.plot()
plt.show()
```
